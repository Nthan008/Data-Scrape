```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://editorial.rottentomatoes.com/guide/all-disney-animated-theatrical-movies-ranked-by-tomatometer/'
url2 = 'https://www.officialcharts.com/charts/singles-chart/'
response = requests.get(url,headers=headers)
response2 = requests.get(url2,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
soup2 = BeautifulSoup(response2.text, "html.parser")

titles = soup.select('div.article_movie_title')
for i in titles:
    print(i)

titles2 = soup2.select('a.chart-name ')
for i in titles2:
    print(i)
```

```python
title = [t.get_text() for t in titles]
for i in title:
    print(i)

title2 = [t.get_text() for t in titles2]
for i in title2:
    print(i)
```

```python
filmtitles=[]
filmTitle = [i[:-13]for i in title]
for i in filmTitle:
    filmtitles.append(i)
    print(i)
    
musictitles = []
musictitle = [j for j in title2]
for j in musictitle:
    musictitles.append(j)
    print(j)
    
```

```python
year = soup.select('span.subtle')
titleyears=[]
title_year = [t.get_text() for t in year]
for i in title_year:
    titleyears.append(i)
    print(i)
```

```python
Artist = soup2.select('a.chart-artist')
Artists=[]
Artist_name = [t.get_text() for t in Artist]
for i in Artist_name:
    Artists.append(i)
    print(i)
```

```python
Week = soup2.select('li.weeks')
Weeks=[]
Week_num = [t.get_text() for t in Week]
Weeknum = [i[7:]for i in Week_num]
for i in Weeknum:
    Weeks.append(i)
    print(i)
```

```python
LastWeek = soup2.select('li.movement')
LastWeeks=[]
Last_Week_num = [t.get_text() for t in LastWeek]
LastWeeknum = [i[4:]for i in Last_Week_num]
LastWeeknumber = [i[:-2]for i in LastWeeknum]
for i in LastWeeknumber:
    LastWeeks.append(i)
    print(i)
```

```python
peak = soup2.select('span.text-brand-cobalt')
Peaks=[]
Peak = [t.get_text() for t in peak]
for i in Peak:
    Peaks.append(i)
    print(i)
```

```python
links = []
link = soup.select('h2 a')
for i in link:
    links.append(i.attrs.get('href'))

for i in links:
    print(i)
```

```python
cast = soup.select('div.cast')
cast_name = [t.get_text() for t in cast]
castnames=[]
castname = [i[11:]for i in cast_name]
for i in castname:
    castnames.append(i)
    print(i)
```

```python
director = soup.select('div.director')
director_name = [t.get_text() for t in director]
directornames=[]
directorname = [i[14:]for i in director_name]
for i in directorname:
    directornames.append(i)
    print(i)
```

```python
    DisneyList = []
    for i in range (len(titles)):
        data = {"title": filmtitles[i],
            "year": titleyears[i],
            "link": links[i],
            "cast": castnames[i],
            "director":directornames[i]}
        DisneyList.append(data)
        
```

```python
df = pd.DataFrame(DisneyList, columns = ['title', 'year', 'link', 'cast', 'director'])
print (df.head())
```

```python
df.to_csv('series.csv', index=False)
pd.read_csv("series.csv", index_col=0)
```
