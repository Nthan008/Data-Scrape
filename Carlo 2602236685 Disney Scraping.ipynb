```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://editorial.rottentomatoes.com/guide/all-disney-animated-theatrical-movies-ranked-by-tomatometer/'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

titles = soup.select('div.article_movie_title')
print(titles[0])
print(" ")
print("how many titles are available?",len(soup.select('div.article_movie_title')))
```

```python
text = soup.select('h2 a')
print(text)

```

```python
title = [t.get_text() for t in titles]
for i in title:
    print(i)
```

```python
filmtitles=[]
filmTitle = [i[:-13]for i in title]
for i in filmTitle:
    filmtitles.append(i)
    print(i)
```

```python
year = soup.select('span.subtle')
print(year)
```

```python
titleyears=[]
title_year = [t.get_text() for t in year]
for i in title_year:
    titleyears.append(i)
    print(i)
```

```python
links = []
for i in text:
    links.append(i.attrs.get('href'))

for i in links:
    print(i)
```

```python
cast = soup.select('div.cast')
print(cast)
```

```python
director = soup.select('div.director')
print(director)
```

```python
cast_name = [t.get_text() for t in cast]
for i in cast_name:
    print(i)
```

```python
castnames=[]
castname = [i[11:]for i in cast_name]
for i in castname:
    castnames.append(i)
    print(i)
```

```python
director_name = [t.get_text() for t in director]
for i in director_name:
    print(i)
```

```python
directornames=[]
directorname = [i[14:]for i in director_name]
for i in directorname:
    directornames.append(i)
    print(i)
```

```python
    DisneyList = []
    for i in range (len(titles)):
        data = {"title": filmtitles[i],
            "year": titleyears[i],
            "link": links[i],
            "cast": castnames[i],
            "director":directornames[i]}
        DisneyList.append(data)
        
```

```python
df = pd.DataFrame(DisneyList, columns = ['title', 'year', 'link', 'cast', 'director'])
print (df.head())
```
